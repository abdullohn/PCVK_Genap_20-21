```
from google.colab import drive
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

original = cv.imread('/content/drive/MyDrive/PCVK-GENAP-2021/female.tiff')
cv2_imshow(original)
```

```
print('Gamma Correction pada citra')
print('----------------------------------')
try: 
  gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
  print('Error, not a number')

gamma_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
     gamma_image[y,x,c] = np.clip(255 * pow(original[y,x,c]/255,(1/gamma)), 0, 255)
     
final_frame = cv.hconcat((original, gamma_image))
cv2_imshow(final_frame)
```

```
img = cv.imread('/content/drive/MyDrive/PCVK-GENAP-2021/teeth.jpg')

mask = np.zeros(img.shape, dtype=np.uint8)
mask = cv.rectangle(mask,pt1=(320,595),pt2=(800,1025),color=(255,255,255),thickness=-2)

result = img & mask

mask1 = cv.rectangle(mask,pt1=(405,5),pt2=(850,455),color=(255,255,255),thickness=-2)

result = img & mask1

final_frame = cv.hconcat((img, result))
cv2_imshow(final_frame)
```
