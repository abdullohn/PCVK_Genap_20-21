```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('/content/drive/MyDrive/PCVK-GENAP-2021/ini.png')

kernel = np.ones((5,5), np.uint8)
dilasi = cv.dilate(img, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Citra Awal')
plt.subplot(122),plt.imshow(dilasi),plt.title('Citra Dilasi')
```
